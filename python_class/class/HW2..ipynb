```python
#Control statements: if, for and while HW
def main():
    heroes = ["iron man", "thor", "spider man", "hulk", "captain america", "hawkeye"]
    print("original heroes:", heroes)
    print("\nif, for and while HW1\n================================================")
    sort_hero(heroes)
    print("================================================\n\nif, for and while HW2\n================================================")
    new_heroes = new(heroes)
    print("================================================\n\nif, for and while HW3\n================================================")
    heroines = man_to_woman(heroes)
    print("================================================\n\nif, for and while HW4\n================================================")
    list_comprehension(heroines)
    print("================================================\n\nif, for and while HW5\n================================================")
    odd(heroes)
    print("================================================\n\nif, for and while HW6\n================================================")
    hero_actors = actors(heroes)
    print("================================================\n\nif, for and while HW7\n================================================")
    match_heroes_actors(new_heroes, hero_actors)
    print("================================================")
    
def sort_hero(heroes):
    # 1. 히어로들을 알파벳 역순으로 정렬하시오.
    heroes.sort()
    heroes.reverse()
    print("1. sorted heroes:", heroes)
    
def new(heroes):
    # 2. 아래 new_heroes에서 첫번째와 마지막 히어로를 제거하고 captain marvel과 black widow을 추가하시오.
    new_heroes = [hero for hero in heroes]
    new_heroes.remove(new_heroes[0])
    new_heroes.remove(new_heroes[-1])
    add_heroes = ["captain marvel", "black widow"]
    for add in add_heroes:
        new_heroes.append(add)
    print("2. change heroes:", new_heroes)
    return new_heroes
    
def man_to_woman(heroes):
    # 3. for문을 이용하여 'man'으로 끝나는 히어로 이름에서 'man'을 'woman'으로 바꾼 새 리스트 heroines를 만드시오.
    heroines = []
    for hero in heroes:
        hero = hero.replace("man", "woman")
        heroines.append(hero)
    print("3. transgender heroines:", heroines)
    return heroines

def list_comprehension(heroines):
    # 4. list comprehension을 이용하여 'woman'이 들어간 히어로 이름을 'human'으로 바꾼 새 리스트 neutralized_heroes를 만드시오.
    neutralized_heroes = [hero.replace("woman", "human") for hero in heroines ]
    print("4. neutralized heroes:", neutralized_heroes)
    
def odd(heroes):
    # 5. heroes에서 enumerate를 이용하여 홀수번째 히어로만 뽑아낸 새 리스트 odd_heroes를 만드시오.
    odd_heroes = []
    for ind, hero in enumerate(heroes):
        if ind % 2 != 0:
            odd_heroes.append(hero)
    print("5.2 odd heroes", odd_heroes)
    
def actors(heroes):
    # 6. for문에 zip 함수를 써서 hero 중 앞의 4명 이름과 배우 이름을 딕셔너리로 묶은 hero_actors를 만들어 보시오.
    actor_names = ["햄식이", "톰 홀랜드", "로다주", "마크 러펄로"]
    hero_actors = dict()
    for hero, actor in zip(heroes, actor_names):
        hero_actors[f"{hero}"] = actor

    print("6. hero actors", hero_actors)
    print("6. thor was played by", hero_actors["thor"])
    print("6. spider man was played by", hero_actors["spider man"])
    return hero_actors
    
def match_heroes_actors(new_heroes, hero_actors):
    # 7. 2의 new_heroes를 연기한 배우들의 이름을 6의 hero_actors를 이용해 출력하시오.
    # 이름이 있으면 6번과 같이 이름을 출력하고 없으면 "?? has no name"을 출력하시오.
    for hero in new_heroes:
        if hero not in hero_actors.keys():
            print(f"{hero} has no name")
        else:
            print(f"{hero} : ",hero_actors[f"{hero}"])
            
if __name__ == '__main__':
    main()
```

```python
#Function HW

def main():
    print("Function HW1 \n==========================")
    list1 = [1,2,3]
    list2 = [5,4,6,8,9]
    input_two_list(list1, list2)
    input_two_list([5,8,7,9], [2,3,4], verbose=True)
    input_two_list([1,2], [7,8,6,2,4,3], shortlen=False, verbose=True)
    input_two_list([11,10,9,8], [5,4,3,2,1,6,8], start=3, verbose=True)
    input_two_list([7,8,5,4,3,2,1], [5,2,1,5,6], shortlen=False, start=2, verbose=True)
    print("==========================\n\nFunction HW2 \n==========================")
    result = average_subjects_kwargs(subject_scores, 10, "cpp", "java", start=1, end=4, skip=[2])
    print(result)
# => result = {'cpp': 580.0, 'java': 800.0}
    result = average_subjects_kwargs(subject_scores, 5, "cpp", "ruby", skip=[3])
    print(result)
    print("==========================")
    
    
#1. 두 개의 숫자 리스트를 입력인자로 받아 원소별로 합산한 리스트를 출력하는 함수를 구현하시오. 함수는 다음과 같은 입력인자를 가져야 한다. 함수를 구현 후 이를 다양하게 사용해보시오.
def input_two_list(list1, list2, shortlen=True, start=0, verbose=False):
    to_list = []
    if shortlen == True:
        if len(list1) > len(list2):
            list1 = list1[:len(list2)]
        elif len(list1) < len(list2):
            list2 = list2[:len(list1)]
        
    elif shortlen == False:
        if len(list1) > len(list2):
            for ind in range(len(list1) - len(list2)):
                list2.append(0)
        
        elif len(list1) < len(list2):
            for ind in range(len(list2) - len(list1)):
                list1.append(0)
                
    if start != 0:
        list1 = list1[start:]
        list2 = list2[start:]
        
    list_range = len(list1)
    
        
    for rang in range(list_range):
        list_add = list1[rang] + list2[rang]
        to_list.append(list_add)
    
    if verbose == True:
        print(to_list)
        
#2위 예시에서 average_multi_subjects() 함수와 average_keyworded_args()를 조합해 입력한 과목들에 multiple 배한 점수의 평균을 start, end, skip, verbose 등의 입력인자를 반영하여 계산하는 함수를 구현하시오.
subject_scores = {"cpp": [57, 36, 80, 53, 23],
                  "java": [46, 88, 72, 15, 54],
                  "ruby": [85, 23, 34, 91, 42]}
def average_list(data, start, end, skip, verbose):
    if end is None:
        avg_data = data[start:]
    else:
        avg_data = data[start:end]

    sum = 0
    for ind, num in enumerate(avg_data):
        if ind not in skip:
            sum += num
    dlen = len(avg_data) - len(skip)
    average = sum / dlen
    if verbose:
        print(f"average {start}~{end} with skipping {skip} = {average}")
    return average

def average_list_with_default(data, start=0, end=None, skip=[], verbose=False):
#     print(data)
    return average_list(data, start, end, skip, verbose)

def average_keyworded_args(data, multiple, **kwargs):
    
    data = [d*multiple for d in data]
    avg = average_list_with_default(data, **kwargs)
    return avg

def average_multi_subjects(scores, *args):
    averages = {}
    for subject in args:
        avg = average_list_with_default(scores[subject], verbose=False)
        averages[subject] = avg
    return averages

def average_subjects_kwargs(scores, multiple, *args, **kwargs):
    result = {}
    sub_avg = average_multi_subjects(scores, *args)
    for sub in args:
        mulit_avg = average_keyworded_args(scores[sub], multiple, **kwargs)
        result[sub] = mulit_avg
    return result


if __name__ == '__main__':
    main()
```

```python
#Class and File IO HW

class MapleStory:
    def __init__(self, name, job):
        self.name = name
        self.job = job
    def status(self):
        print("name:", self.name)
        print("job:", self.job)
    def skill(self):
        print("skill list:", None)

class Warrior(MapleStory):
    def __init__(self, name, job):
        super().__init__(name, job)
    def skill(self):
        print("skill list: power slash")
    def quest(self):
        print("Quest: help warrior!!")
        
class Magician(MapleStory):
    def __init__(self, name, job):
        super().__init__(name, job)
    def skill(self):
        print("skill list: magic claw")
    def quest(self):
        print("Quest: help magician!!")

        
def write_txt(matrix_str):
    with open("matrix.txt", "w") as f:
        f.write(matrix_str)
        
def con_matrix():
    with open("matrix.txt", 'r') as f:
        matrix = f.readlines()
        matrix = [line.rstrip("\n") for line in matrix]
    print(matrix)
    matrix = [line.split(" ") for line in matrix]
    print(matrix)
    matrix_int = []
    for line in matrix:
        line = [int(num) for num in line]
        print(line)
        matrix_int.append(line)
    print(matrix_int)
    matrix = matrix_int
    return matrix

def determinant(matrix):
    deter = matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0]
    print("determinant:", deter)
    
    
def main():
    print("Class HW1\n=============================")
    game = [MapleStory("joon", "novice"), Warrior("sweet", "warrior"), Magician("judy", "magician")]
    for gam in game:
        gam.status()
        gam.skill()
        print("\n")
    war = Warrior("ohland", "warrior")
    war.status()
    war.skill()
    war.quest()
    
    print("=============================\n\nClass HW2\n=============================")
    matrix_str = "4 2\n5 3"
    write_txt(matrix_str)
    matrix = con_matrix()
    determinant(matrix)
    print("=============================")
if __name__ == '__main__':
    main()
```
